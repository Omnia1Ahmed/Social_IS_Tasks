```python
import itertools
import random
import networkx as nx
```

```python
def barabasi_albert_graph(n,m):
    G = nx.complete_graph(m + 1)
    for i in range(G.number_of_nodes(), n):
        new_nibors = []
        possip_nibors = list(G.nodes)
        for _ in range(m):
            degrees = [G.degree(nod) for nod in possip_nibors]
            j = random.choices(possip_nibors, degrees)[0]
            new_nibors.append(j)
            possip_nibors.remove(j)
            
        for j in new_nibors:
            G.add_edge(i,j)
    return G
```

```python
G=barabasi_albert_graph(30,1)
nx.draw(G, with_labels=True, node_color = 'LightCoral', node_size = 800, font_color ='white' )
```
