```python
import networkx as nx
%matplotlib inline
```

```python
G = nx.Graph()

G.add_nodes_from(['a','b','c','d'])

G.add_edges_from([('a','b'),('b','c'),('a','c'),('a','d')])

nx.draw(G, with_labels=True , node_size = 2000, node_color="#D4AE7C")
```

```python
nx.has_path(G, 'c', 'd')
```

```python
list(nx.all_simple_paths(G,'c', 'd'))
```

```python
nx.shortest_path(G, 'c', 'd')
```

```python
nx.shortest_path_length(G, 'c', 'd')
```

```python
nx.is_connected(G)
```

```python
C = nx.Graph()

nx.add_cycle(C,(1,2,3))
C.add_edge(4,5)

nx.draw(C, with_labels=True, node_size = 2000, node_color="#ff8aae" )
```

```python
nx.is_connected(C)
```

```python
nx.has_path(C, 3, 5)
```

```python
nx.number_connected_components(C)
```

```python
list(nx.connected_components(C))
```

```python
comp = list(nx.connected_components(C))
len(comp[0])
```

```python
max(nx.connected_components(C), key=len)
```

```python
core_n = max(nx.connected_components(C), key=len)
core = C.subgraph(core_n)

nx.draw(core, with_labels=True,node_size = 2000, node_color="#fff89a")
```

```python
Di = nx.DiGraph()
Di.add_edges_from([
    (1,2),
    (2,3),
    (3,2), (3,4), (3,5),
    (4,2), (4,5), (4,6),
    (5,6),
    (6,4),
])
nx.draw(Di, with_labels=True,node_size = 2000, node_color="#b2a6ff")
```

```python
nx.has_path(Di, 1, 6)
```

```python
nx.has_path(Di, 4, 1)
```

```python
nx.shortest_path(Di, 5, 2)
```

```python
nx.is_strongly_connected(Di)
```

```python
nx.is_weakly_connected(Di)
```

```python
list(nx.weakly_connected_components(Di))
```

```python
list(nx.strongly_connected_components(Di))
```
